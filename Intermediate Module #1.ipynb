### Comparing Lists and Tuples

```python
### comparing the size
import sys
my_list = [0, 1, 2, "hello", True]
my_tuple = (0, 1, 2, "hello", True)
print(sys.getsizeof(my_list), "bytes")
print(sys.getsizeof(my_tuple), "bytes")

## the output shows tuple takes smaller size than list
```

```python
### comparing time
import timeit
print(timeit.timeit(stmt="[0, 1, 2, 3, 4, 5]", number=100000)) ## 1m times to create a list
print(timeit.timeit(stmt="(0, 1, 2, 3, 4, 5)", number=100000)) ## 1m times to create a tuple

### so list takes longer time when creation than tuple
```

```python
### Calculating Union, and Intersection of sets
odds = {1, 3, 5, 7, 9}
evens = {2, 4, 6, 8}
primes = {2, 3, 5, 7}
### union of odds and ebens
u = odds.union(evens)
u
```

```python
### we cal calculate intersection
I = odds.intersection(evens) ### intersection of odds and evens
print(I)
```

```python
### we cal calculate intersection odds and primes
i = odds.intersection(primes)
print(i)
```

```python
## we can calculate difference
setA = {1, 2, 3, 4, 5, 6, 7, 8, 9}
setB = {1, 2, 3, 10, 11, 12}
diff = setA.difference(setB) ### elements found in setA but not in setB
diff
```

```python
diff = setB.difference(setA) ### elements found in setB but not in setA
diff
```

```python
### symmetric Difference
diff = setA.symmetric_difference(setB) 
diff
```

```python
## Updating
setA.update(setB) ## updating setA by adding the elements from setB(without duplicating) 
print(setA)
```

```python
setA.intersection_update(setB) ## updating setA by the common elements
print(setA)
```
